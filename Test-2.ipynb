```python
def plot_bar(feature, value, title, df, size):
    f, ax = plt.subplots(1,1, figsize=(2*size,7))
    df = df.sort_values([value], ascending=False).reset_index(drop=True)
    g = sns.barplot(df[feature][0:], df[value][0:], palette='ch:.05')
    g.set_title(" ".format(title))
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    plt.show()
```

```python
plot_bar("NAME","Case_Total","Total Cases",data, size=10)
```

```python
plot_bar("NAME","Recovered","Total Recovered",data, size=10)
```

```python
plot_bar("NAME","Deaths","Total Deaths",data, size=10)
```

```python
plot_bar("NAME","Tests","Total Tests Performed",data, size=10)
```
